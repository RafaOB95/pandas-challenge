```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset. 
#merge ->(1 data_set, 2 data_set, how I want to merge)
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

District Summary

```python
school_data_complete['school_name'].value_counts()
total_school = school_data_complete['school_name'].nunique()
print('The total number of unique schools is: ' + str(total_school))
```

```python
school_data_complete['Student ID'].value_counts()
total_students = len(school_data_complete)
print('The total number of students are: ' + str(total_students))
```

```python
school_data_complete['budget'].value_counts()
budget = school_data_complete['budget'].unique()
budget_sum = budget.sum()
print('The total budget is: $' + str(budget_sum))
```

```python
school_data_complete['math_score'].value_counts()
math = school_data_complete['math_score'].sum()
math_av=math/len(school_data_complete)
print('The average in Math is: ' + str(math_av))
```

```python
school_data_complete['reading_score'].value_counts()
read = school_data_complete['reading_score'].sum()
read_av=read/len(school_data_complete)
print('The average in Reading is: ' + str(read_av))
```

```python
# Use the following to calculate the percentage of students who passed math
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / len(school_data_complete) * 100
print('The percentage of students who passed math: ' + str(passing_math_percentage) + ' %')
```

```python
# Use the following to calculate the percentage of students who passed reading
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / len(school_data_complete) * 100
print('The percentage of students who passed reading: ' + str(passing_reading_percentage) + ' %')
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  len(school_data_complete) * 100
print('The percentage of students who passed math and reading: ' + str(overall_passing_rate) + ' %')
```

```python
# initialize data of lists.
data = {'Number of unique schools': [total_school],
        'Total students': [total_students],
       'Total budget': [budget_sum],
       'Average math score': [math_av],
       'Average reading score': [read_av],
       '% passing math': [passing_math_percentage],
       '% passing reading': [passing_reading_percentage],
       '% overall passing': [overall_passing_rate],}
  
# Create DataFrame
district_summary = pd.DataFrame(data)

# Formatting
district_summary["Total students"] = district_summary["Total students"].map("{:,}".format)
district_summary["Total budget"] = district_summary["Total budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
a = ninth_graders.groupby(['school_name'])['math_score'].sum()/ninth_graders['school_name'].value_counts()
b = tenth_graders.groupby(['school_name'])['math_score'].sum()/tenth_graders['school_name'].value_counts()
c = eleventh_graders.groupby(['school_name'])['math_score'].sum()/eleventh_graders['school_name'].value_counts()
d = twelfth_graders.groupby(['school_name'])['math_score'].sum()/twelfth_graders['school_name'].value_counts()

math_scores_by_grade = pd.concat([a, b, c, d], axis=1)
math_scores_by_grade.rename(
    columns={0: "9th", 1: "10th", 2: "11th", 3: "12th"},
    inplace=True,
)

math_scores_by_grade
```

```python
a = ninth_graders.groupby(['school_name'])['reading_score'].sum()/ninth_graders['school_name'].value_counts()
b = tenth_graders.groupby(['school_name'])['reading_score'].sum()/tenth_graders['school_name'].value_counts()
c = eleventh_graders.groupby(['school_name'])['reading_score'].sum()/eleventh_graders['school_name'].value_counts()
d = twelfth_graders.groupby(['school_name'])['reading_score'].sum()/twelfth_graders['school_name'].value_counts()

reading_scores_by_grade = pd.concat([a, b, c, d], axis=1)
reading_scores_by_grade.rename(
    columns={0: "9th", 1: "10th", 2: "11th", 3: "12th"},
    inplace=True,
)

reading_scores_by_grade
```
